```python
import scipy.stats.qmc as qmc
import numpy as np
 
def create_lhs(d, N, seed, safe=False):
    """
    Create a Latin Hypercube Sampling (LHS) of parameter trajectories.

    Parameters:
    - d (int): Number of dimensions.
    - N (int): Number of samples.
    - seed (int): Seed for random number generation.

    Returns:
    - parameter_trajectories (ndarray): Array of parameter trajectories.

    """ 
 
    # create latin hypercube with N samples and d dimensions
    sampler = qmc.LatinHypercube(d, seed=seed)
    parameter_trajectories = np.array(sampler.random(n=N))
       
    # save LHS in file
    if safe:
        np.save("lhs.npy", parameter_trajectories)
 
    return parameter_trajectories
```

```python
import matplotlib.pyplot as plt


lhs = create_lhs(d=2, N=100, seed=1)

plt.scatter(lhs[:, 0], lhs[:, 1])
plt.xlabel('Parameter 1')
plt.ylabel('Parameter 2')
plt.title('Scatter Plot of LHS Numbers')
plt.show()
```
